### Importation of libraries and definition of functions.

```python
import os
import otbApplication as otb
def combination_fusion(input_pan_image, input_ms_image, output_fusion_image, method="rcs"):
    # Crea la aplicación Pansharpening
    app = otb.Registry.CreateApplication("Pansharpening")

    # Establece los parámetros de entrada
    app.SetParameterString("inxs", input_ms_image)
    app.SetParameterString("inp", input_pan_image)
    app.SetParameterString("method", method)
    #app.SetParameterString("ram", 10240)
    if method == "rcs":
        app.SetParameterString("method.rcs.radiusx", "9")
        app.SetParameterString("method.rcs.radiusy","9")
    elif method == "lmvm":
        app.SetParameterString("method.lmvm.radiusx", "3")
        app.SetParameterString("method.lmvm.radiusy", "3")
    else:
        app.SetParameterString("method.bayes.lambda", "0.999")
        app.SetParameterString("method.bayes.s", "1")

    # Establece el parámetro de salida
    app.SetParameterString("out", output_fusion_image)
    app.SetParameterOutputImagePixelType("out", 2)

    # Ejecuta la aplicación
    app.ExecuteAndWriteOutput()

def fusion_bands_M_H(S2_bands_folder, PS_bands_folder, output_folder, nbands, nbands2):
    # Carpeta para las bandas fusionadas
    output_fusion_folder = os.path.join(output_folder, "S2_f")
    
    # Verificar si la carpeta de salida existe, si no, crearla
    if not os.path.exists(output_fusion_folder):
        os.makedirs(output_fusion_folder)
    if nbands == 8 and nbands2 == 12:
        band_pairs = {
            "band2": "band2",
            "band3": "band3",
            "band4": "band6",
            "band5": "band7",
            "band6": "band7",
            "band7": "band7",
            "band8": "band8",
            "band10": "band8",
            "band11": "band11_est",
            "band12": "band12_est"
        }
    elif nbands == 5 and nbands2 == 11:
        band_pairs = {
            "band2": "band1",
            "band3": "band2",
            "band4": "band3",
            "band5": "band5",
            "band6": "band10_est",
            "band7": "band11_est",
        }
    elif nbands == 4 and nbands2 == 11:
        band_pairs = {
            "band2": "band1",
            "band3": "band2",
            "band4": "band3",
            "band5": "band4",
            "band6": "band10_est",
            "band7": "band11_est"
        }
    elif nbands == 5 and nbands2 == 7:
        band_pairs = {
            "band1": "band1",
            "band2": "band2",
            "band3": "band3",
            "band4": "band4",
            "band5": "band5_est",
            "band6": "band7_est"
        }  
    elif nbands == 4 and nbands2 == 12:
        band_pairs = {
            "band2": "band1",
            "band3": "band2",
            "band4": "band3",
            "band5": "band34_avg",
            "band6": "band34_avg",
            "band7": "band34_avg",
            "band8": "band4",
            "band10": "band4",
            "band11": "band11_est",
            "band12": "band12_est"
        }

    # Realizar la fusión con los tres métodos y guardar las imágenes en la carpeta de salida
    for s2_band, ps_band in band_pairs.items():
        input_pan_image = os.path.join(PS_bands_folder, f"{ps_band}.tif")
        input_ms_image = os.path.join(S2_bands_folder, f"{s2_band}.tif")
        #["rcs", "lmvm", "bayes"]
   
        for method in ["bayes"]:
            output_fusion_image = os.path.join(output_fusion_folder, f"{s2_band}_{method}_fusion.tif")
            
            # Realizar la fusión utilizando el método especificado
            combination_fusion(input_pan_image, input_ms_image, output_fusion_image, method)
```
